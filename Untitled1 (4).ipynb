```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import shutil # Import shutil for potential cleanup

# Dataset parameters
DATA_DIR = 'food-101/images'  # Main images directory
TARGET_SIZE = (224, 224)      # EfficientNet input size (can be changed to (380,380) for EfficientNetB4)
BATCH_SIZE = 32
VAL_SPLIT = 0.2               # Validation split percentage
SEED = 42                     # Random seed for reproducibility

# Create separate train/val directories
train_dir = 'food-101/train'
val_dir = 'food-101/val'

# Clean up existing train/val directories before creating new ones
if os.path.exists(train_dir):
    print(f"Cleaning up existing directory: {train_dir}")
    shutil.rmtree(train_dir)
os.makedirs(train_dir, exist_ok=True)

if os.path.exists(val_dir):
    print(f"Cleaning up existing directory: {val_dir}")
    shutil.rmtree(val_dir)
os.makedirs(val_dir, exist_ok=True)


# Copy files for train/val split instead of creating symbolic links
print("Organizing train/val split by copying files...")
for class_name in os.listdir(DATA_DIR):
    class_dir = os.path.join(DATA_DIR, class_name)
    if os.path.isdir(class_dir):
        # Create class directories in train and val
        os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
        os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)

        # Get all images and split
        images = [f for f in os.listdir(class_dir) if f.endswith('.jpg')]
        split_idx = int(len(images) * (1-VAL_SPLIT))

        # Copy training data
        print(f"Copying training files for class: {class_name}")
        for img in images[:split_idx]:
            src = os.path.join(class_dir, img)
            dst = os.path.join(train_dir, class_name, img)
            if os.path.exists(src):
                try:
                    shutil.copyfile(src, dst)
                except IOError as e:
                    print(f"Error copying file {src} to {dst}: {e}")
            else:
                print(f"Warning: Source file not found for copying: {src}")


        # Copy validation data
        print(f"Copying validation files for class: {class_name}")
        for img in images[split_idx:]:
            src = os.path.join(class_dir, img)
            dst = os.path.join(val_dir, class_name, img)
            if os.path.exists(src):
                 try:
                    shutil.copyfile(src, dst)
                 except IOError as e:
                    print(f"Error copying file {src} to {dst}: {e}")
            else:
                 print(f"Warning: Source file not found for copying: {src}")


# Data augmentation configuration for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Validation data generator (no augmentation)
val_datagen = ImageDataGenerator(rescale=1./255)

# Create generators from separate directories
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=TARGET_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True,
    seed=SEED
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=TARGET_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False  # Important for validation
)

# Verify the generators
print("\nTraining classes:", train_generator.class_indices)
print("Training samples:", train_generator.samples)
print("Validation samples:", val_generator.samples)

# Suggested next steps:
# 1. Build your model (EfficientNetB0/B4 recommended)
# 2. Train using the generators
# model.fit(train_generator, validation_data=val_generator, epochs=10)
```

```python
# Initialize the EfficientNetB4 base model with pre-trained ImageNet weights
base_model = EfficientNetB4(
    weights='imagenet',         # Load weights pre-trained on ImageNet dataset
    include_top=False,          # Exclude the original classification head (top layers)
    input_shape=(224, 224, 3)   # Specify input tensor shape (height, width, channels)

    # Additional available parameters (using defaults here):
    # pooling=None,            # Pooling mode for feature extraction (None, 'avg', 'max')
    # classes=1000,            # Number of output classes (only relevant if include_top=True)
    # classifier_activation='softmax'  # Activation function for top layer
)

```

```python
# Extract the output tensor from the base model
# This will be a 4D tensor with shape (batch_size, height, width, channels)
x = base_model.output

# Add a Global Average Pooling layer to reduce spatial dimensions
# Converts each feature map to a single value by taking average of all spatial values
# Output shape: (batch_size, channels)
x = GlobalAveragePooling2D(name='global_avg_pool')(x)

# Add a fully-connected (Dense) layer with ReLU activation
# This helps learn complex patterns from the pooled features
# 1024 units is a good starting point - can be adjusted based on dataset size
x = Dense(1024, activation='relu', name='fc1')(x)

# Add Dropout layer for regularization
# Randomly sets 50% of input units to 0 during training to prevent overfitting
x = Dropout(0.5, name='dropout')(x)

# Final output layer with softmax activation
# Number of units matches number of classes in our dataset
# Softmax converts outputs to class probabilities (sum to 1)
predictions = Dense(
    len(train_generator.class_indices),
    activation='softmax',
    name='output'
)(x)

# Create the complete model by connecting input and output
model = Model(
    inputs=base_model.input,  # Input tensor from base model
    outputs=predictions,      # Our custom output layers
    name='food_classifier'
)

# Compile the model with appropriate settings
model.compile(
    optimizer=tf.keras.optimizers.Adam(
        learning_rate=0.001,  # Default learning rate
        beta_1=0.9,          # Exponential decay rates for moment estimates
        beta_2=0.999
    ),
    loss='categorical_crossentropy',  # Standard for multi-class classification
    metrics=[
        'accuracy',  # Overall accuracy
        tf.keras.metrics.TopKCategoricalAccuracy(k=3, name='top3_accuracy')  # Also track top-3 accuracy
    ]
)

# Improved recommendations:
# 1. Consider learning rate scheduling for better convergence:
#    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
#        initial_learning_rate=0.001,
#        decay_steps=10000,
#        decay_rate=0.9)
#    optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

# 2. For imbalanced datasets, consider class weights:
#    class_weights = compute_class_weight(...)
#    Then pass to model.fit(..., class_weight=class_weights)

# 3. Add more metrics for better evaluation:
#    metrics=[
#        'accuracy',
#        tf.keras.metrics.Precision(name='precision'),
#        tf.keras.metrics.Recall(name='recall'),
#        tf.keras.metrics.AUC(name='auc')
#    ]
```

```python
# Food Recognition, Portion Estimation, and Calorie Calculation System
# Google Colab Implementation

import cv2
import numpy as np
from PIL import Image
import requests
from io import BytesIO
from google.colab import files
import matplotlib.pyplot as plt
import pandas as pd

class FoodNutritionEstimator:
    # Food database with approximate calories per ml
    FOOD_DATABASE = {
        'vegetables': 0.4,    # kcal/ml
        'fruits': 0.6,        # kcal/ml
        'grains': 1.2,        # kcal/ml
        'protein': 1.5,       # kcal/ml
        'dairy': 0.8,         # kcal/ml
        'fats': 8.0           # kcal/ml
    }

    def __init__(self):
        self.food_categories = {
            'vegetables': {'color_range': ([35, 50, 50], [85, 255, 255]), 'density': 0.8},
            'fruits': {'color_range': ([10, 100, 50], [25, 255, 255]), 'density': 0.9},
            'grains': {'color_range': ([20, 50, 50], [30, 255, 255]), 'density': 1.1},
            'protein': {'color_range': ([0, 50, 50], [20, 255, 255]), 'density': 1.3},
            'dairy': {'color_range': ([0, 0, 100], [180, 50, 255]), 'density': 1.0},
            'fats': {'color_range': ([15, 50, 50], [25, 255, 255]), 'density': 0.9}
        }

    def load_image(self, image_path):
        """Load image from path or URL with enhanced error handling"""
        try:
            if image_path.startswith(('http://', 'https://')):
                response = requests.get(image_path, timeout=10)
                response.raise_for_status()
                img = Image.open(BytesIO(response.content))
                print("Image successfully loaded from URL")
                return np.array(img)
            else:
                img = Image.open(image_path)
                print("Image successfully loaded from local path")
                return np.array(img)
        except Exception as e:
            print(f"Error loading image: {str(e)}")
            return None

    def preprocess_image(self, img):
        """Convert and standardize image format"""
        if len(img.shape) == 2:  # Grayscale
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        elif img.shape[2] == 4:  # RGBA
            img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)
        else:  # RGB
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        return img

    def detect_food_regions(self, img):
        """Identify food regions using multi-color segmentation"""
        hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
        combined_mask = np.zeros((img.shape[0], img.shape[1]), dtype=np.uint8)
        category_masks = {}

        for category, params in self.food_categories.items():
            lower, upper = params['color_range']
            mask = cv2.inRange(hsv, np.array(lower), np.array(upper))
            category_masks[category] = mask
            combined_mask = cv2.bitwise_or(combined_mask, mask)

        return combined_mask, category_masks

    def estimate_portion_size(self, img, show_visualization=False):
        """Estimate food portion size and nutritional content"""
        try:
            # Preprocess image
            img = self.preprocess_image(img)
            height, width = img.shape[:2]

            # Calculate reference dimensions (plate detection)
            plate_diameter_pixels = min(height, width) * 0.7
            plate_diameter_inches = 10  # Standard plate size
            pixels_per_inch = plate_diameter_pixels / plate_diameter_inches

            # Detect food regions
            combined_mask, category_masks = self.detect_food_regions(img)
            total_food_area = np.sum(combined_mask > 0)

            # Calculate volume and calories by category
            results = {}
            total_volume = 0
            total_calories = 0

            for category, mask in category_masks.items():
                area = np.sum(mask > 0)
                if area == 0:
                    continue

                # Calculate volume (area * average height * density factor)
                area_inches = area / (pixels_per_inch ** 2)
                volume_ml = area_inches * 1.5 * 16.387 * self.food_categories[category]['density']
                calories = volume_ml * self.FOOD_DATABASE[category]

                results[category] = {
                    'volume_ml': round(volume_ml, 1),
                    'calories': round(calories, 1),
                    'percentage': round(100 * area / total_food_area, 1)
                }

                total_volume += volume_ml
                total_calories += calories

            # Visualization
            if show_visualization:
                self.visualize_results(img, combined_mask, category_masks)

            return {
                'total_volume_ml': round(total_volume, 1),
                'total_calories': round(total_calories, 1),
                'category_breakdown': results,
                'pixels_per_inch': round(pixels_per_inch, 2)
            }

        except Exception as e:
            print(f"Estimation error: {str(e)}")
            return None

    def visualize_results(self, img, combined_mask, category_masks):
        """Display processing results with matplotlib"""
        plt.figure(figsize=(15, 10))

        # Original image
        plt.subplot(2, 2, 1)
        plt.imshow(img)
        plt.title("Original Image")
        plt.axis('off')

        # Combined food mask
        plt.subplot(2, 2, 2)
        plt.imshow(combined_mask, cmap='gray')
        plt.title("All Food Regions")
        plt.axis('off')

        # Category masks
        plt.subplot(2, 2, 3)
        category_vis = np.zeros_like(img)
        colors = plt.cm.get_cmap('tab10', len(category_masks))

        for i, (category, mask) in enumerate(category_masks.items()):
            category_vis[mask > 0] = np.array(colors(i))[:3] * 255

        plt.imshow(category_vis)
        plt.title("Food Categories")
        plt.axis('off')

        # Masked food
        plt.subplot(2, 2, 4)
        masked_img = cv2.bitwise_and(img, img, mask=combined_mask)
        plt.imshow(masked_img)
        plt.title("Detected Food")
        plt.axis('off')

        plt.tight_layout()
        plt.show()

def main():
    print("Food Nutrition Estimation System")
    print("==============================")

    estimator = FoodNutritionEstimator()

    # Get user input
    print("\n1. Upload an image file")
    print("2. Use sample image URL")
    choice = input("Enter choice (1/2): ")

    if choice == '1':
        uploaded = files.upload()
        if not uploaded:
            print("No file uploaded. Using sample image instead.")
            image_url = "https://images.unsplash.com/photo-1546069901-ba9599a7e63c"
            img = estimator.load_image(image_url)
        else:
            image_path = next(iter(uploaded.keys()))
            img = estimator.load_image(image_path)
    else:
        image_url = "https://images.unsplash.com/photo-1546069901-ba9599a7e63c"
        img = estimator.load_image(image_url)

    if img is not None:
        print("\nAnalyzing food content...")
        results = estimator.estimate_portion_size(img, show_visualization=True)

        if results:
            print("\n=== Estimation Results ===")
            print(f"Total Volume: {results['total_volume_ml']} ml")
            print(f"Total Calories: {results['total_calories']} kcal")
            print("\nCategory Breakdown:")

            df = pd.DataFrame.from_dict(results['category_breakdown'], orient='index')
            print(df)
        else:
            print("Failed to analyze image.")
    else:
        print("Failed to process image.")

if __name__ == "__main__":
    main()
```
